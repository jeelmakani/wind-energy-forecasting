```python
valid[["wind_generation_actual", "Forecast_DT"]].plot(figsize=(14, 7))
```

```python
rf_rmse = np.sqrt(mean_squared_error(valid["wind_generation_actual"], valid.Forecast_DT))
rf_mae = mean_absolute_error(valid["wind_generation_actual"], valid.Forecast_DT)

print("RMSE of Random Forests:", rf_rmse)
print("\nMAE of Random Forests:", rf_mae)
```
